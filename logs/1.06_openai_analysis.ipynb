```python
def ask_gpt(model, prompt, timeout=1):
    print(f"asking: {prompt}")
    sleep(timeout)
    gpt_response = openai.ChatCompletion.create(model=model, messages=[{"role": "user", "content": prompt}])   
    return gpt_response.choices[0].message.content
```

```python
#load clustered and filtered dataset

data = pd.read_csv('tob_domain_only.csv')
data.head()
```

```python
# how many for each cluster?
cluster_counts = data.value_counts('question_cluster')
cluster_counts
```

```python
import math

sample_size = cluster_counts.apply(lambda x: math.ceil(x * .1))
```

```python
sample = pd.DataFrame({'full': cluster_counts, 'sample': sample_size})
```

```python
sample
```

```python
def ask_cluster_questions(raw, cluster, sample_size):
    print(f"Fetching {sample_size} question(s) from cluster {cluster}")
    input_sample = raw[raw['question_cluster'] == cluster].sample(sample_size)
    input_sample = input_sample[['question_cluster', 'question']].rename({'question_cluster': 'cluster'}, axis=1)
    
    input_sample['answer:gpt-3.5-turbo-16k'] = input_sample["question"].apply(lambda q: ask_gpt('gpt-3.5-turbo-16k', q, .2))
    
    return input_sample
```

```python
ask_cluster_questions(data, 118, 1)
```

```python
gpt_df = pd.DataFrame({'cluster': [], 'question': [], 'answer:gpt-3.5-turbo-16k': []}, index=[], dtype=int)

for cluster, count_data in sample.iterrows():
    cluser_answers = ask_cluster_questions(data, cluster, count_data['sample'])
    
    gpt_df = pd.concat([gpt_df, cluser_answers])

    sleep(10)
    
gpt_df
```

```python
gpt_df
```

```python
joined = sample.join(gpt_df.value_counts('cluster'))

new_sample = joined[joined['count'].isnull()]
```

```python
new_sample
```

```python
gpt_df
```

```python
gpt_df2 = gpt_df.copy()

for cluster, count_data in new_sample.iterrows():
    cluser_answers = ask_cluster_questions(data, cluster, int(count_data['sample']))
    
    gpt_df2 = pd.concat([gpt_df2, cluser_answers])

    sleep(10)
    
gpt_df2
```

```python
gpt_df2.to_csv('gpt_answers_by_cluster_sample.csv')
```
