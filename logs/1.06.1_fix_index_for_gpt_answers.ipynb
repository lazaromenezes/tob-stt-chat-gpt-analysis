```python
# Needs to fix indexing. The index column for gpt_answers_by_cluster_sample.csv ended up being the row count index for tob_domain_only.csv
# This notebook fixes it by replacing it to the proper value. 

import pandas as pd

tob_data = pd.read_csv('./tob_domain_only.csv')

tob_data
```

```python
gpt_data = pd.read_csv('./gpt_answers_by_cluster_sample.csv')

gpt_data['tob_index'] = gpt_data['Unnamed: 0'].apply(lambda x: tob_data.iloc[x]['index'])

gpt_data
```

```python
gpt_data = gpt_data[['tob_index', 'cluster', 'question', 'answer:gpt-3.5-turbo-16k']]

gpt_data
```

```python
gpt_data = gpt_data.rename({'tob_index': 'index'}, axis=1)
```

```python
gpt_data = gpt_data.set_index('index')
```

```python
gpt_data.to_csv('./gpt_answers_sample.csv')
```
