```python
def exclude_from_domain(df, cluster):
    mask = df['question_cluster'] == cluster
    df.loc[mask, 'is_domain'] = False
```

```python
def show_n_most_frequent(df, label, number = 20):
    most_frequent = df.loc[df['is_domain']].value_counts(label, normalize=True).nlargest(number)
    for cluster, value in most_frequent.items():
        print(f"cluster #: {cluster}: {df[df['question_cluster'] == cluster].head(1)['question'].values[0]}")
```
