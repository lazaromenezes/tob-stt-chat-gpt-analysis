```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

tob_final = pd.read_csv('./tob_v1_and_v2.csv', index_col='index')
```

```python
encoder = SentenceTransformer('all-mpnet-base-v2')
```

```python
tob_copy = tob_final.copy()
```

```python
def compare_answers(row):
    v1_embedings = encoder.encode(row['answer_v1'])
    v2_embedings = encoder.encode(row['answer_v2'])

    cosine_similarity = util.cos_sim(v1_embedings, v2_embedings)

    return cosine_similarity.item()

tob_copy['v1_x_v2'] = tob_copy.apply(compare_answers, axis=1)
```
